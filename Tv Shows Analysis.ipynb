```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('seaborn-deep')
```

## Data Pre-Processing
#### The Dataset provided is quite cleanexcept for Nan values and index values.In this we have eliminated the inconsistencies.

```python
shows_df=pd.read_csv('shows.csv')
shows_df.rename(columns = {'Unnamed: 0':'S.No'}, inplace = True)
shows_df['Age']=shows_df['Age'].fillna('All')
shows_df['IMDb']=shows_df['IMDb'].fillna(4.5)
shows_df['Rotten Tomatoes']=shows_df['Rotten Tomatoes'].fillna('25%')
shows_df['Rotten Tomatoes'] = shows_df['Rotten Tomatoes'].str.replace('\%','').astype('float')
shows_df=shows_df.drop('S.No',axis=1)
shows_df.reset_index()
shows_df.index.name='S.No'
shows_df
```

## Data Arrangement
#### Making copies of the data into different datasets. Each dataset has been sorted based on different constraints

### Sorting Based on Year

```python
years_df=shows_df.copy()
years_df=years_df.sort_values(by='Year')
years_df
```

### Sorting Based on IMDb Rating

```python
imdb_df=shows_df.copy().sort_values(by='IMDb',ascending=False).reset_index()
imdb_df=imdb_df.drop('S.No',axis=1)
imdb_df
```

### Sorting Based on Rotten Tomatoes Rating

```python
rt_df=shows_df.sort_values(by='Rotten Tomatoes',ascending=False).reset_index()
rt_df=rt_df.drop('S.No',axis=1)
rt_df
```

### Data Frame Containing Only Netflix Data

```python
netflix_df=shows_df.loc[shows_df['Netflix'] == 1]
netflix_df=netflix_df.reset_index()
netflix_df=netflix_df.drop(['S.No', 'Netflix','Hulu', 'Prime Video', 'Disney+', 'type'],axis=1)
netflix_df
```

### Data Frame Containing Only Hulu Data

```python
hulu_df=shows_df.loc[shows_df['Hulu'] == 1]
hulu_df=hulu_df.reset_index()
hulu_df=hulu_df.drop(['S.No', 'Netflix','Hulu', 'Prime Video', 'Disney+', 'type'],axis=1)
hulu_df
```

### Data Set Containing Only Prime Video Data

```python
prime_df=shows_df.loc[shows_df['Prime Video'] == 1]
prime_df=prime_df.reset_index()
prime_df=prime_df.drop(['S.No', 'Netflix','Hulu', 'Prime Video', 'Disney+', 'type'],axis=1)
prime_df
```

### Data Frame Containing Disney+ Data

```python
dp_df=shows_df.loc[shows_df['Disney+'] == 1]
dp_df=dp_df.reset_index()
dp_df=dp_df.drop(['S.No', 'Netflix','Hulu', 'Prime Video', 'Disney+', 'type'],axis=1)
dp_df
```

### Graph Showing the No.of Shows in each OTT Platform with an IMDb rating >=8.5

```python
netflix_df['>=8.5']=netflix_df['IMDb']>8.4
hulu_df['>=8.5']=hulu_df['IMDb']>8.4
prime_df['>=8.5']=prime_df['IMDb']>8.4
dp_df['>=8.5']=dp_df['IMDb']>8.4
c1=netflix_df['>=8.5'].values.sum()
c2=hulu_df['>=8.5'].values.sum()
c3=prime_df['>=8.5'].values.sum()
c4=dp_df['>=8.5'].values.sum()
ir=[]
ir.append(c1)
ir.append(c2)
ir.append(c3)
ir.append(c4)
```

```python
ir
```

```python
great_rate=pd.Series(ir,index=['Netflix', 'Hulu', 'Prime Video', 'Disney+'])
great_rate
```

```python
plt.figure()
sns.barplot(x=great_rate.index,y= great_rate, palette='magma').set_title('Shows having above 8.4 rating in IMDb')
```

### Graph Showing the No.of  Shows in each OTT Platform with a Rotten Tomatoes>=85%

```python
netflix_df['>=85.0']=netflix_df['Rotten Tomatoes']>84.0
hulu_df['>=85.0']=hulu_df['Rotten Tomatoes']>84.0
prime_df['>=85.0']=prime_df['Rotten Tomatoes']>84.0
dp_df['>=85.0']=dp_df['Rotten Tomatoes']>84.0
r1=netflix_df['>=85.0'].values.sum()
r2=hulu_df['>=85.0'].values.sum()
r3=prime_df['>=85.0'].values.sum()
r4=dp_df['>=85.0'].values.sum()
tr=[]
tr.append(r1)
tr.append(r2)
tr.append(r3)
tr.append(r4)
```

```python
tr
```

```python
great_trate=pd.Series(tr,index=['Netflix', 'Hulu', 'Prime Video', 'Disney+'])
great_trate
```

```python
plt.figure()
sns.barplot(x=great_trate.index,y= great_trate, palette='magma').set_title('Shows having above 84% rating in Rotten Tomatoes')
```

```python
labels=['Netflix', 'Hulu', 'Prime Video', 'Disney+']
cir_plot=plt.Circle((0,0), 0.7, color='white')
```

```python
plt.figure()
plt.pie(great_trate, labels=labels)
p=plt.gcf()
p.gca().add_artist(cir_plot)
plt.title('Distribution of Shows Having Rotten Tomatoes Rating >84%')
plt.show()
```
